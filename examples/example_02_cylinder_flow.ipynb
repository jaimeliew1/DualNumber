```python
import numpy as np
import dualitic as dl
import matplotlib.pyplot as plt
```

```python
def u(x, y, U=1, a=1, C=0):
    return U * (1 - a**2 * (x**2 - y**2) / ((x**2 + y**2) ** 2)) - C / (
        2 * np.pi
    ) * y / (x**2 + y**2)


def v(x, y, U=1, a=1, C=0):
    return -2 * U**2 * a**2 * x * y / ((x**2 + y**2) ** 2) + C / (
        2 * np.pi
    ) * x / (x**2 + y**2)
```

```python
a = 1
C = dl.DualNumber(5, [1])

x = np.linspace(-5, 5, 1000)
y = np.linspace(-5, 5, 1000)

xmesh, ymesh = np.meshgrid(x, y)

U = u(xmesh, ymesh, a=a, C=C)
V = v(xmesh, ymesh, a=a, C=C)

mask = (xmesh**2 + ymesh**2) < 1

U[mask] = np.nan
V[mask] = np.nan
E = 0.5 * (U**2 + V**2)
```
